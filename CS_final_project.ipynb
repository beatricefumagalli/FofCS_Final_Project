###### We want to evaluate if movie_id can be used as unique index

```python
def id (df, column):
    if len(df[column]) == len(df[column].drop_duplicates()):
        return "true"
id(crd,'movie_id') and id(mv,'movie_id')
```

###### Now we can define a function to extract the dataframe written in format json and then remove the original columns

```python
def ext(df, column):
    ris = df[column].apply(pd.read_json) 
    movie_id = df['movie_id'] 
    for j in range(len(df)):
        ris[j]['movie_id'] = movie_id[j] 
    return pd.concat(ris.values).reset_index().drop("index", axis = 1)

cast=ext(crd,'cast')
crew=ext(crd,'crew')
genres=ext(mv,'genres')
keywords=ext(mv,'keywords')
production_companies=ext(mv,'production_companies')
production_countries=ext(mv,'production_countries')
spoken_languages=ext(mv,'spoken_languages')

movies= mv.drop(['genres', 'keywords', 'production_companies', 'production_countries', 'spoken_languages'],axis = 1)

credits= crd.drop(['cast', 'crew'], axis = 1)
```

###### 4) Extract the domain of each homepage

```python
def extract_domain(each_homepage):
    domains = []
    homepage = each_homepage.split(" ")
    for j in range(len(homepage)):
        domains.append(homepage[j].split('/')[2].split('.')[-1])
    return str(domains)[1:-1].replace("\'", "")
mv2["domains"] = mv2["homepage"].apply(extract_domain)
domain_of_each_homepage = mv2[["title", "homepage", "domains"]]
domain_of_each_homepage
```

###### 10) For each company and each decade, compute the overall revenue

```python
movies['year'] = pd.DatetimeIndex(movies['release_date']).year
movies['year'] = np.array(movies['year'], dtype=int)

#to obtain the decade from 'str(year)', we take the first three values of the year with the command [:3] and then we add a '0' at the end 

def ext_decade(year):
    return str(year)[:3] + '0'
movies['decade'] = movies['year'].apply(ext_decade)

movies_companies_id = production_companies.merge(right = movies[['movie_id','revenue','decade']], on = 'movie_id')
movies_companies_name = production_companies_name.merge(right = movies_companies_id[['company_id','revenue','decade']], on = 'company_id')
overall_companies_revenue = movies_companies_name.groupby(by = ['company_id','name','decade'], as_index = False).sum()
overall_companies_revenue
```

###### 11) For each decade, compute the company with maximum revenue

```python
max_revenue_each_decade = overall_companies_revenue.groupby(['decade'], as_index=False)['revenue'].max()
company_max_revenue_each_decade = overall_companies_revenue.merge(max_revenue_each_decade, on='revenue').sort_values(by=['decade_x']).drop(['decade_x'], axis=1).rename(columns = {'decade_y':'decade'})
company_max_revenue_each_decade
```

###### 13) Distribute the revenue according to the order of appearance in a movie. Assume that the i-th actor contributes twice as much as the (i+1)-th actor to the revenue.

```python
revenue = movies[['movie_id', 'title', 'revenue']]
rev_actor = pd.merge(cast, revenue, on='movie_id')
rev_actor['order'] = rev_actor['order'].astype(int)

#Actor in position 0 must be twice than actor in position 1, actor in position 1 must be twice than actor in position 2, ecc..

rev_actor['num'] = 0.5**(rev_actor['order']) #0.5 raised to the corresponding 'order'
den = rev_actor.groupby('movie_id')['num'].sum().to_frame('den').reset_index() #to divide in equal parts from 0 to 1
rev_actor = pd.merge(rev_actor, den, on='movie_id')

rev_actor.head()
```

```python
rev_actor['% of revenue'] = rev_actor['num']/rev_actor['den']

rev_actor['rev_actor'] = rev_actor['% of revenue']*rev_actor['revenue']

rev_actor
```

```python
pd.set_option('display.float_format', lambda x: '%.0f' % x)
rev_actor.groupby(['movie_id', 'title']).agg({'revenue':'first', 'rev_actor':'sum'})
```

###### 14) For each actor find the total revenue attributed to him/her

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
revenue_actor=rev_actor.groupby(['actor_id','name'])['rev_actor'].sum().to_frame('Total revenue')
revenue_actor
```
